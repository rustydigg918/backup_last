```python
from urllib.request import urlretrieve
urlretrieve(URL, 'iris.csv')
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 25, 30
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
plt.figure(figsize=(62,32))
df = pandas.read_csv("iris.csv")

print(df) 
```

```python
df.Name.unique()
```

```python
d = {'Iris-setosa': 1, 'Iris-versicolor': 0, 'Iris-virginica': 2}
df['Name'] = df['Name'].map(d)
```

```python
features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalLength']

X = df[features]
y = df['Name']

print(X)
print(y) 
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')

img=pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show() 
```
