# Clustering Analysis
### Clustering is the task of dividing the population or data points into a number of groups such that data points in the same groups are more similar to other data points in the same group than those in other groups. In simple words, the aim is to segregate groups with similar traits and assign them into clusters.

### Types of Clustering
-- Broadly speaking, clustering can be divided into two subgroups :

   #### Hard Clustering: In hard clustering, each data point either belongs to a cluster completely or not. For example, in the above example each customer is put into one group out of the 10 groups.
    
   #### Soft Clustering: In soft clustering, instead of putting each data point into a separate cluster, a probability or likelihood of that data point to be in those clusters is assigned. For example, from the above scenario each costumer is assigned a probability to be in either of 10 clusters of the retail store.

```python
# ------------------------------------------------------------------------#
#                        Types of CLustering Algorithms
#                     -----------------------------------
# Since the task of clustering is subjective, the means that can be used for 
# achieving this goal are plenty. Every methodology follows a different set 
# of rules for defining the ‘similarity’ among data points. In fact, there 
# are more than 100 clustering algorithms known. But few of the algorithms 
# are used popularly, let’s look at them in detail:
# ------------------------------------------------------------------------#
```

```python
#####################
# Connectivity models:
#####################
# these models are based on the notion that the data points closer in data 
# space exhibit more similarity to each other than the data points lying 
# farther away. These models can follow two approaches. In the first approach, 
# they start with classifying all data points into separate clusters & then 
# aggregating them as the distance decreases. In the second approach, all data 
# points are classified as a single cluster and then partitioned as the distance 
# increases. Also, the choice of distance function is subjective. These models 
# are very easy to interpret but lacks scalability for handling big datasets. 
# Examples of these models are hierarchical clustering algorithm and its variants.
```

```python
#####################
# Centroid models:
#####################
# These are iterative clustering algorithms in which the notion of similarity 
# is derived by the closeness of a data point to the centroid of the clusters. 
# K-Means clustering algorithm is a popular algorithm that falls into this category. 
# In these models, the no. of clusters required at the end have to be 
# mentioned beforehand, which makes it important to have prior knowledge 
# of the dataset. These models run iteratively to find the local optima.
# Example- K-Means Clustering
```

```python
#####################
# Distribution models:
#####################
# These clustering models are based on the notion of how probable is it that 
# all data points in the cluster belong to the same distribution 
# (For example: Normal, Gaussian). These models often suffer from overfitting. 
# A popular example of these models is Expectation-maximization algorithm 
# which uses multivariate normal distributions.
```

```python
#####################
# Density Models:
#####################
# These models search the data space for areas of varied density of data points 
# in the data space. It isolates various different density regions and assign 
# the data points within these regions in the same cluster. Popular examples 
# of density models are DBSCAN and OPTICS.
```

# K-Means Clustering

```python
# Import necessary libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
from IPython.display import Image
```

```python
# Import iris data
iris = datasets.load_iris()
iris_data = iris.data
```

```python
#------------------------#
# The K-Means Algorithm
#------------------------#
```

```python
# Step:1
# Let there be N data points. At first, K 'random' centroids are initialised in 
# our data set representing K different clusters. 
# Example: If k = 2, assign some data points to cluster-1 (red)
           # and remaining to cluster-2 (Grey)
Image('kmeans_step1.png')
```

```python
# Step:2 (Cluster Assignment Step)
# Now, each of the N data points are assigned to closest centroid in 
# the data set and merged with that centroid as a single cluster. 
# In this way, every data point is assigned to one of the closest centroids.
# Notice below that the red dots are assigned to the red cross, same for grey
Image('kmeans_step2.png')
```

```python
# Step:3 (Centroid Calculation Step- This moves centroids)
# Reassign the data points that are closest to the centroids
# Then recalculate the centroids (mean of the closest points) 
# so new centroids will be formed. i.e. centroids are moved to the mean of clusters
# Step-2 and Step-3 are iterative process till centroids do not change anymore
# Refer to the following figure
Image('kmeans_step3.png')
```

```python
# Notice how the clusters are getting formed
```

```python
# Keep in mind that the default distance for k-Means is "Euclidean Distance"
# However there are so many other distance metrices such as:
        # Manhattan
        # cosine
        # Jaccard
        # chebyshev
        # mahalanobis
Image('Euclidean_Manhattan.png')
```

```python
# Run KMeans Clustering
KMeans_model = KMeans(n_clusters=3, init='random')
KMeans_model.fit(iris_data)
clusters = KMeans_model.predict(iris_data)
```

```python
# DETERMINING THE NUMBER OF CLUSTERS
####################################
# How do you choose k? There isn't a bright line, but we can evaluate 
# performance metrics such as the silhouette coefficient and 
# Elbow method for within sum of squared errors across values of k.
```

```python
# Create a bunch of different models
k_rng = range(1,7)
KMeans_model_many = [KMeans(n_clusters = k).fit(iris_data) for k in k_rng]
```

```python
#================================================================
# Option 1: Silhouette Coefficient (SC)
# Generally want SC to be closer to 1, while also minimizing k
#================================================================

# It is a measure of how close each point in one cluster is 
# to points in the neighboring clusters. 
# This measure has a range of [-1, 1].
```

```python
# Calculate Silhouette Score
from sklearn import metrics
silhouette_score = [metrics.silhouette_score(iris_data, km.labels_, metric='euclidean') \
                    for km in KMeans_model_many[1:]]
```

```python
# Plot the results
plt.figure(figsize=(7, 8))
plt.subplot(211)
plt.plot(k_rng[1:], silhouette_score, 'b*-')
plt.xlim([1,7])
plt.grid(True)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')
plt.plot(3,silhouette_score[1], 'o', markersize=12, markeredgewidth=1.5,
         markerfacecolor='None', markeredgecolor='r')
```

```python
# -----------------------------------------------------|
#                 ELBOW METHOD                         |
#      Calculate the within sum of squared errors 
#           and plot over a range of k
# -----------------------------------------------------

# Recall that, the basic idea behind partitioning methods, such as 
# k-means clustering, is to define clusters such that the total intra-cluster 
# variation [or total within-cluster sum of square (WSS)] is minimized. 
```

```python
#===================================================================
# Option 2: Within Sum of Squares(WSS) [a.k.a., inertia or Distortion]
# Generally want to minimize WSS, while also minimizing k
#===================================================================
```

```python
# Wss or 'inertia' is already calculated and available in your model
within_sum_squares = [km.inertia_ for km in KMeans_model_many]
```

```python
# Plot the sum of squared errors results
plt.plot(k_rng, within_sum_squares, 'b*-')
plt.xlim([1,7])
plt.grid(True)
plt.title('Using the elbow method to inform k choice')
plt.xlabel('k number of clusters')
plt.ylabel('Within Sum of Squares')
plt.plot(3,within_sum_squares[2], 'ro', markersize=12, markeredgewidth=1.5,
         markerfacecolor='None', markeredgecolor='r')
```

```python
# K-Means Vs K-Means++
######################

# Actually there is no K-Means Vs K-Means++
# K-means++ is just an initialization procedure for K-means.

# Usual K-means with 'random' initialization starts with allocating cluster 
# centers (as many k you specified) randomly and then looks for "better" solutions. 
# K-means++ starts with allocation one cluster 
# center randomly and then searches for other centers given the first one. 
# So both algorithms use random initialization as a starting point, 
# so can give different results on different runs.
```

# Hierarchical Clustering

```python
# Some basics on Hierarchical Clustering
###########################################

# It is a type of connectivity model clustering which is based on the fact 
# that data points that are closer to each other are more similar than the 
# data points lying far away in a data space.

# 

# As the name speaks for itself, the hierarchical clustering forms the 
# hierarchy of the clusters that can be studied by visualising dendogram.
```

```python
#####################################
# Types of Hierarchical Clustering
######################################

# Agglomerative Clustering: 
# ------------------------#
# Also known as bottom-up approach or hierarchical 
# agglomerative clustering (HAC). A structure that is more informative than 
# the unstructured set of clusters returned by flat clustering. 
# This clustering algorithm does not require us to prespecify the number 
# of clusters. Bottom-up algorithms treat each data as a singleton cluster 
# at the outset and then successively agglomerates pairs of clusters until 
# all clusters have been merged into a single cluster that contains all data.

# Divisive clustering :
# ---------------------#
# Also known as top-down approach. 
# This algorithm also does not require to prespecify the number of clusters. 
# Top-down clustering requires a method for splitting a cluster that contains 
# the whole data and proceeds by splitting clusters recursively until 
# individual data have been splitted into singleton cluster.
```

```python
# The Algorithm:
# 1. Let there be N data points. Firstly, these N data points are assigned 
    # to N different clusters with one data point in each cluster.
    # i.e. Data points are assigned to their own clusters

# 2. Then, two data points with minimum euclidean distance between them 
    # are merged into a single cluster. [Or any other distance]

# 3. Then, two clusters with minimum centroid distance between them 
    # are merged into a single cluster.

# 4. This process is repeated until we are left with a single cluster, 
    # hence forming hierarchy of clusters.
```

```python
# First step is to find the number of clusters
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(iris_data, 'single')
plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top'
)
plt.show()  
```

# Good Cluster Analysis

```python
# Data-points within same cluster share similar profile: 
#-------------------------------------------------------#
# Statistically, check the standard deviation for each input variable 
# in each cluster. A perfect separation in case of cluster analysis is 
# rarely achieved. Hence, even one standard deviation distance between 
# two cluster means is considered to be a good separation.

    
# Well spread proportion of data-points among clusters: 
#------------------------------------------------------#
# There are no standards for this requirement.
# But a minimum of 5% and a maximum of 35% of the total population can 
# be assumed as a safe range for each cluster.
```

# Revisiting STEPs to perform Cluster Analysis

```python
# Step:1- Hypothesis building
###############################
# This is the most crucial step of the whole exercise. 
# Try to identify all possible variables that can help segment the 
# portfolio regardless of its availability
```

```python
# Step:2- Initial shortlist of variable 
#######################################
# Once we have all possible variable, start selecting variable as per the data availability.
```

```python
# Step:3- Visualize the data
###############################
# It is very important to know the population spread across the selected 
# variable before starting any analysis. It will most likely give you some hints
# on the cluster formations in a visual way. If that happens, you will definitely
# see that reflected in final clusters
```

```python
# Step:4- Data cleaning and Transformation
###########################################
#  Cluster analysis is very sensitive to outliers. It is very important to 
# clean data on all variables taken into consideration. 

# 1. Remove the outliers : 
# (Not recommended in case the total data-points are low in number) 
# We remove the data-points beyond mean +/- 3*standard deviation.

# 2. Capping and flouring of variables : 
# (Recommended approach) We cap and flour all data-points at 1 and 99 percentile.

# Feature scaling. This is very important. Typically realword data are all
# in different units and scale. In such scenarios, the variables with higher
# values dominate the 'distance' calculations and may lead to poor cluster formations
```

```python
# Step:5- Variable clustering
###############################
# This does not really mean clustering of the variables
# It means that, if there are related variables (e.g. high multicollinearity),
# You must include one of them and drop the rest of the related ones
# Thumrule is, minimum the number of variables, better it is
# When you have a lot of important variables, you might have to
# perform Principal Component Analysis (PCA) and cluster with limited variables
```

```python
# Step:6- CLUSTERING
######################
# You cluster with the most appropriate clustering algorithm
# If the clusters are not well separated, Hierarchical might 
# make more sense compared to K Means
```

```python
# Step:7- Evaluating Clustering
###############################
# See the within cluster Sum of Squares (WCSS) against the number of clusters

# check Data-points within same cluster share similar profile. A Statistical 
# method to judge this criterion is simply checking the standard deviation 
# for each input variable in each cluster.

# Well spread proportion of data-points among clusters
# As a thumb rule, a good cluster analysis has all clusters with population
# between 5-30% of the overall base. Say, my total number of customer 
# for bank X is 10000. The minimum and maximum size of any cluster should 
# be 500 and 3000. If any of the cluster is beyond the limit than repeat 
# the procedure with additional number of variables.
```

```python
# Step:8- Profiling the clusters
###############################
# Note that this also helps in manual evaluation of the clusters
# There are certain data points that you can manually see and find if 
# they should be in same clusters or different clusters

#  Then, you need to identify behavior of each cluster. 
# Example: Lets say we map age and income to each of the four 
# clusters and get following results:
Image('cluster_profiling_example.png')
```
