```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_excel('Emp_data.xlsx', "Sheet1")
df.head(5)
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
fig = plt.figure()
fig.add_subplot(111)
plt.scatter(x,y)
plt.show()
```

```python
fig = plt.figure()
fig.add_subplot(1,2,1)
fig.add_subplot(2,2,2)
fig.add_subplot(2,2,4)
plt.show()
```

```python
fig = plt.figure()
viz = fig.add_subplot(1,1,1)
viz .hist(df['Age'],bins =4)

plt.title('Age distribution by bin')
plt.xlabel('Age')
plt.ylabel('No of employee in each bin')
plt.show()
```

```python
fig = plt.figure()
viz = fig.add_subplot(1,1,1)
viz.boxplot(df['Age'])
plt.title('Boxplot for Age')
plt.show()
```

```python
fig = plt.figure()
viz = fig.add_subplot(1,1,1)
viz.scatter(df['Age'], df['Sales'])
plt.title('Scatter plot for Age')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['Age'], df['Sales'], s=df['Income'])
plt.title('Scatter plot for Age')
plt.show()
```

```python
#barplot
A = [5.,30.,45.,22.]
B = [5.,25.,50.,20.]
X = range(4)

plt.bar(X,A,color = 'r')
```

```python
#barplot
A = [5.,30.,45.,22.]
B = [5.,25.,50.,20.]
X = range(4)

plt.bar(X,A,color = 'g')

plt.bar(X,B, color = 'r', bottom = A)
plt.show()
```

```python
labels = 'Frog', 'Hog', 'Dog', 'Log'
sizes  = [15,30,45,10]
explode =(0, 0.1 ,0 ,0)

figl,axl = plt.subplots()
axl.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow = True, startangle=90)
axl.axis('equal')
plt.show()
```

```python
import seaborn as sns
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
sns.distplot(iris["SepalWidth"])
```

```python
sns.distplot(df['Age'], kde=False, bins =4)
```

```python
sns.distplot(iris["SepalWidth"], kde = False, bins =15)
```

```python
import scipy.stats as stats
sns.distplot(iris["SepalLength"], kde = False, bins =25, fit = stats.gamma)
```

```python
sns.jointplot(x="SepalLength", y = "SepalWidth", data=iris, height=5)
plt.show()
```

```python
sns.FacetGrid(iris, hue="Name", height=5) \
.map(plt.scatter, "SepalLength", "SepalWidth") \
.add_legend()
plt.show()
```

```python
sns.boxplot(x="Name", y ="PetalLength", data=iris)
plt.show()
```

```python
ax = sns.boxplot(x="Name",y="PetalLength", data=iris)
ax = sns.stripplot(x="Name",y="PetalLength", data=iris, jitter=True,edgecolor="gray")
plt.show()
```

```python
sns.violinplot(x="Name",y="PetalLength", data=iris, size=6)
plt.show()
```

```python
sns.FacetGrid(iris, hue="Name", height=5) \
.map(sns.kdeplot, "PetalLength") \
.add_legend()
plt.show()
```

```python
sns.pairplot(iris,height=3)
plt.show()
```

```python
sns.pairplot(iris,height=3, diag_kind="kde")
plt.show()
```

```python
sns.pairplot(iris,hue="Name", height =3)
plt.show()
```
