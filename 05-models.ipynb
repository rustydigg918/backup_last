```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
from sklearn import linear_model
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X=tips['total_bill'], y=tips['tip'])
```

```python
lr.fit(X=tips['total_bill'].values.reshape(-1, 1), y=tips['tip'])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.fit(X=tips[['total_bill', 'size']], y=tips['tip'])
```

```python
lr.coef_
```

```python
tips
```

```python
# dummy encoding
# one-hot encoding
```

```python
lr.fit(X=tips[['total_bill', 'sex']], y=tips['tip'])
```

```python
tips_dummy = tips[['tip', 'total_bill', 'sex']]
```

```python
tips_dummy.head()
```

```python
tips_dummy = pd.get_dummies(tips_dummy, drop_first=True)
```

```python
tips_dummy.head()
```

```python
lr.fit(X=tips_dummy.iloc[:, 1:], y=tips_dummy.iloc[:, 0])
```

```python
tip_money = lr.coef_[0]
```
